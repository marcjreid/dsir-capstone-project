# Problem Statement

When I was young I was a fan of the show Pokemon. Pokemon cards cards are fun to collect and are even be a good investment as some rare cards sell for thousands of dollars. I'm interesting in finding the best way for someone to upload their pokemon card collection online. This could be a fast way to  record someone's collection, or even be made compatible with a Pokemon computer game. The question is, can we make a system that can accurately classify Pokemon images so that video or pictures of pokemon will let software know what pokemon a collector has? My apporach is to use a classification model using nueral networks. The features will be different pictures of each Pokemon found online, and the classes are a Pokemon's number in the Pokedex (Bulbasaur is number 1 and so on).

# Preprocessing

```python
# Use PIL to open images and convert all to RGB (dimensions will be different otherwise)
# (May later try converting all images to greyscale to improve results)
pika_goog1 = Image.open(r"./simple_images/Pikachu/Pikachu_1.jpeg").convert('RGB')
pika_goog2 = Image.open(r"./simple_images/Pikachu/Pikachu_2.jpeg").convert('RGB')
sqrtl_goog1 = Image.open(r"./simple_images/Squirtle/Squirtle_1.png").convert('RGB')
sqrtl_goog2 = Image.open(r"./simple_images/Squirtle/Squirtle_2.png").convert('RGB')
```

```python
img1 = img1.convert('RGB')
img2 = img2.convert('RGB')
sq_img1 = sq_img1.convert('RGB')
sq_img2 = sq_img2.convert('RGB')
```

```python
sq_img1 = sq_img1.convert('L')
sq_img1 = sq_img1.convert('L')
```

```python
# Create train and test array with images arrays as inputs, and Pokedex number as labels
# train = np.array([google_pic_array, class_num])
# test = np.array([tcg_pic_array, class_num])
train = np.array([[np.array(pika_goog1.resize((256,256))), 25], 
                  [np.array(pika_goog2.resize((256,256))), 25], 
                  [np.array(sqrtl_goog1.resize((256,256))), 7], 
                  [np.array(sqrtl_goog2.resize((256,256))), 7]])
test = np.array([[np.array(img1.resize((256,256))), 25], 
                [np.array(img2.resize((256,256))), 25], 
                [np.array(sq_img1.resize((256,256))), 7], 
                [np.array(sq_img2.resize((256,256))), 7]])
```

```python
# Populate X train with pokemon image and label information
x_train = []
y_train = []
for pokemon, number in train:
  x_train.append(pokemon)
  y_train.append(number)
```

```python
x_test = []
y_test = []
for pokemon, number in test:
  x_test.append(pokemon)
  y_test.append(number)
```

```python
# Strandarize image data
x_train = np.array(x_train) / 255
x_test = np.array(x_test) / 255
```

```python
# Reshape array. Keras model expects 4 dimensional input data
x_train.reshape(-1, 256, 256, 1)
x_test.reshape(-1, 256, 256, 1)
```

```python
# Reshape label array. Keras expects 2 dimensional y data
y_train = np.asarray(y_train).astype('float32').reshape((-1,1))
y_test = np.asarray(y_train).astype('float32').reshape((-1,1))
```

```python
# Check to see shape of feature and label data has been changed successfully
x_train[2].shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Simple Classification Model

```python
#Simple Keras nueral network with two hidden layers and convulutional filters to classify images
model = Sequential()
model.add(Conv2D(16, (3, 3), activation ='relu', input_shape=(256, 256, 3)))
model.add(Conv2D(16, (3, 3), activation = 'relu'))
model.add(MaxPool2D((2,2)))

model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile model with binary cross entropy and accruacy as this is a binary classification problem now.
model.compile(loss='bce', metrics=['acc'])
```

```python
# fit model to training and testing data for 5 epochs.
history = model.fit(x_train, y_train, validation_data = (x_test, y_test), epochs=5)
```

```python
# Loss for training and validation fall at similar rates
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
```

```python
# Accuracy for training and validation at 0
plt.plot(history.history['acc'], label='Train Accuracy')
plt.plot(history.history['val_acc'], label='Validation Accuracy')
plt.legend()
```

The accruacy is 0. I beleive the model and input data are incompatible in some way. I will run second simple model with different layers an label names.

```python
train2 = np.array([[np.array(pika_goog1.resize((256,256))), 1], 
                  [np.array(sqrtl_goog1.resize((256,256))), 0]])

test2 = np.array([[np.array(pika_goog2.resize((256,256))), 1], 
                [np.array(sqrtl_goog1.resize((256,256))), 0]])
```

```python
x_train2 = []
y_train2 = []
for feature, label in train2:
  x_train2.append(feature)
  y_train2.append(label)
```

```python
x_test2 = []
y_test2 = []
for feature, label in test2:
  x_test2.append(feature)
  y_test2.append(label)
```

```python
x_train2 = np.array(x_train2) / 255
x_test2 = np.array(x_test2) / 255
```

```python
x_train2.reshape(-1, 256, 256, 1)
x_test2.reshape(-1, 256, 256, 1)
```

```python
y_train2 = np.asarray(y_train2).astype('float32').reshape((-1,1))
y_test2 = np.asarray(y_train2).astype('float32').reshape((-1,1))
```

```python
#Simple Keras nueral network with two hidden layers and convulutional filters to classify images
model2 = Sequential()
model2.add(Conv2D(16, (3, 3), activation ='relu', input_shape=(256, 256, 3)))
model2.add(Flatten())
model2.add(Dense(100, activation='relu'))
model2.add(Dense(100, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))
model2.compile(loss='categorical_crossentropy', metrics=['acc'])
```

```python
history2 = model2.fit(x_train2, y_train2, validation_data = (x_test2, y_test2), epochs=5)
```

```python
plt.plot(history2.history['acc'], label='Train Accuracy')
plt.plot(history2.history['val_acc'], label='Validation Accuracy')
plt.legend()
```
